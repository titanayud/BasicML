# 1 - Support Vector Classification

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
data = pd.read_csv('archive/Mall_Customers.csv')
```

```python
Age = data['Age']
Spending_Score= data['Spending Score (1-100)']
Gender = data['Gender']
```

```python
df = pd.DataFrame(dict( Age=Age, Spending_Score=Spending_Score, Gender=Gender))
fig, ax = plt.subplots()
colors = {'Male':'blue', 'Female':'red'}
ax.scatter(df['Age'], df['Spending_Score'], c=df['Gender'].map(colors))

plt.show()
```

```python
df.info()
```

```python
X = df[df.columns[:2]]
Y = df ['Gender']
```

```python
from sklearn.preprocessing import StandardScaler
 
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
```

```python
from sklearn.model_selection import train_test_split
 
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.33, random_state=42)
```

```python
from sklearn.svm import SVC

clf = SVC()
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```
